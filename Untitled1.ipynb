```python
%matplotlib inline
import pandas as pd
import numpy 
import re
from datetime import datetime
TAG_RE = re.compile(r'<[^>]+>')
def post(m):
    return {"thread":m.group(1),
            "post":m.group(2),
            "author":m.group(3),
            "date":datetime.strptime(m.group(4), '%B %d, %Y %I:%M %p'),
            "content":TAG_RE.sub('',m.group(5))
            }
def parse(html):
    return [post(m) for m in re.finditer(r"\<dt\>Thread\:\<\/dt\>\s*<dd>(.*?)</dd>"+
                         r".*?\"\>(.*?)\<\/a\>"+
                         r".*?\"profileCardAvatarThumb\"\>\s*(.*?)\s*\<\/span\>"+
                         r".*?\<dd\>\s*(.*?)\<\/dd\>"+
                         r".*?\"dbThreadBody\"\>(.*?)\<div class\=\"dbThreadFooter\"\>", html, flags=re.DOTALL)]
with open("threads/Collection01.webarchive") as f:
    allpost = parse(f.read())
```

```python
from collections import Counter
counter = Counter(x["date"].weekday() for x in allpost)

for post in allpost:
    post["weekday"] = post["date"].weekday()
    post["length"] = len(post["content"])
    post["minutes_of_the_day"] = post["date"].hour * 60 +post["date"].minute



c = pd.DataFrame(allpost)
with open("threads/allpost.csv", "w+") as f:
    f.write(c.to_csv())
```
